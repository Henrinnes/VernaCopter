```python
# assume waypoints shape = (2, N)
x = waypoints[0, ::5]
y = waypoints[1, ::5]

# differences between consecutive points
dx = np.diff(x)
dy = np.diff(y)

# plot points/trajectory
plt.plot(x, y, 'o-', label="trajectory")

# plot arrows
plt.quiver(
    x[:-1], y[:-1],   # start points
    dx, dy,           # vector components
    angles='xy', scale_units='xy', scale=1,
    color='r'
)

plt.axis('equal')
plt.legend()
plt.show()
```
